# Intergrals and Derivates

```python
x= np.linspace(1,10,100)
y = 1/x**2 * np.sin(x)

plt.plot(x,y)
```

```python
np.cumsum(np.array([1,2,3,4]))
```

```python
y_int = np.cumsum(y) * (x[1]-x[0])
dydx = np.gradient(y,x)
plt.plot(x,dydx)
plt.plot(x,y)
plt.plot(x,y_int)
```

###  Reorganizing Arrays

```python
before = np.array([[1,2,3,4],[5,6,7,8]])
print(before)

after = before.reshape((4,2))
print(after)
```

####  Vertically stacking vectors

```python
v1 = np.array([1,2,3,4])
v2 = np.array([5,6,7,8])

np.vstack([v1,v2,v1])
```
